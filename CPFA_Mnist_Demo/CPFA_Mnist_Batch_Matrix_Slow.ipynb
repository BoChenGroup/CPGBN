```python
#!/usr/bin/env python2
# -*- coding: utf-8 -*-
"""
Created on Wed Jan 10 22:41:31 2018

@author: wangchaojie
"""

import numpy as np
np.random.RandomState(1)

realmin = 2.2e-308
def log_max(x):
    return np.log(np.maximum(x, realmin))

#====================== Load data ======================#
import scipy.io as sio
data = sio.loadmat('./mnist_gray')
train_data= np.array(np.ceil(data['train_mnist']*25), order ='C')[:,0:10]   #0-1
test_data = np.array(np.ceil(data['test_mnist']*25),  order ='C')[:,0:10]   #0-1

N = train_data.shape[1]
V1 = 28
V2 = 28

#====================== Setting ======================#
Setting = {}                                        # an empty dict
Setting['N']    = N                                 # Batch Number
Setting['K']    = 16                                # Kernel Number
Setting['V1']   = V1
Setting['V2']   = V2                                # Input Size
Setting['K3']   = 28 
Setting['K4']   = 3                                 # Kernel Size
Setting['K1']   = Setting['V1'] + 1 - Setting['K3'] # Feature Map size 
Setting['K2']   = Setting['V2'] + 1 - Setting['K4']
Setting['Iter'] = 500

#====================== SuperParams ======================#
SuperParams = {}
SuperParams['gamma0'] = 0.1
SuperParams['c0']     = 0.1
SuperParams['eta']    = 0.05

#====================== Initial ======================#
Params = {}
Params['D'] = np.random.rand(Setting['K3'], Setting['K4'], Setting['K'])
Params['W'] = np.random.rand(Setting['K1'], Setting['K2'], Setting['K'], Setting['N'])

import time
Iter_time = []
Iter_lh   = []

#====================== Gibbs ======================#
import Conv_PFA_Sampler as CPS
from scipy.special import gamma

for my_iter in range(Setting['Iter']):
    
    start_time = time.time()
    
    #====================== Augmentation ======================#
    X_vnk = np.zeros([Setting['V1'] * Setting['V2'], Setting['N'], Setting['K']],order = 'C')
    X_vnk_params = np.zeros([Setting['V1'] * Setting['V2'], Setting['N'], Setting['K']],order = 'C')  # vnk
    
    for k in range(Setting['K']):
        D_k     = Params['D'][:,:,k]  / Params['D'][:,:,k].sum()
        D_k_Aug = CPS.Conv_Aug(D_k,np.array([Setting['K1'],Setting['K2']]))
        W_k_sn  = np.reshape(np.transpose(Params['W'][:,:,k,:],[2,1,0]),[Setting['N'],Setting['K1']*Setting['K2']]).T
        X_vnk_params[:,:,k] = np.dot(D_k_Aug,W_k_sn)
    
    for n in range(Setting['N']):
        X_vk_params = np.array(X_vnk_params[:,n,:],order = 'C')
        tmp, X_vnk[:,n,:] = CPS.Multrnd_Matrix(np.reshape(train_data[:,n],[train_data[:,n].shape[0],1],order = 'C').astype('double'), X_vk_params , np.ones([Setting['K'],1], order = 'C') )

    #====================== Parameters Update ======================#
    for k in range(Setting['K']):
        
        # Update D
        X_34 = np.zeros([Setting['K3']*Setting['K4'], 1], order = 'C')
        for n in range(Setting['N']):
            W_kn_Aug = CPS.Conv_Aug(Params['W'][:,:,k,n],np.array([Setting['K3'],Setting['K4']]))
            D_k = np.reshape(np.transpose(Params['D'][:,:,k],[1,0]),[Setting['K3']*Setting['K4'],1])
            X_34n,tmp = CPS.Multrnd_Matrix(np.array(np.reshape(X_vnk[:,n,k], [X_vnk[:,n,k].shape[0], 1]), order = 'c'),np.array(W_kn_Aug, order = 'c')  , np.array(D_k, order = 'c'))
            X_34 = X_34 + X_34n
        
        D_k = (X_34 + SuperParams['eta'] ) / (X_34 + SuperParams['eta'] ).sum()
        Params['D'][:,:,k] = np.transpose(np.reshape(D_k,[Setting['K4'], Setting['K3']]), [1,0])
        
        # Update W
        D_k = Params['D'][:,:,k]  / Params['D'][:,:,k].sum()
        D_k_Aug = CPS.Conv_Aug(D_k,np.array([Setting['K1'],Setting['K2']]))
        W_k_sn = np.reshape(np.transpose(Params['W'][:,:,k,:],[1,0,2]),[Setting['K1']*Setting['K2'],Setting['N']])
        X_k_12n, tmp = CPS.Multrnd_Matrix(np.array(X_vnk[:,:,k], order = 'C'), np.array(D_k_Aug, order = 'C')  , np.array(W_k_sn, order = 'C') )
    
        W_k_sn = np.random.gamma(SuperParams['gamma0'] + X_k_12n) / (1 + 0)
        Params['W'][:,:,k,:] = np.transpose(np.reshape(W_k_sn, [Setting['K2'], Setting['K1'],Setting['N']]), [1,0,2])
        
    end_time = time.time()
    
    if my_iter == 0:
        Iter_time.append(end_time - start_time)
    else:
        Iter_time.append(end_time - start_time + Iter_time[-1])
    
    #====================== Likelihood ======================#
    lamda = np.sum(X_vnk_params,2)    
    k = train_data
    likelyhood = np.sum(k * log_max(lamda) - lamda - log_max(gamma(k + 1)))   / Setting['N']

    Iter_lh.append(likelyhood)
    print likelyhood, (end_time - start_time)
    
import matplotlib.pyplot as plt
plt.plot(Iter_time,Iter_lh)
plt.show()
```
