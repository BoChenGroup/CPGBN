```python
import codecs
import copy
import numpy as np

elec_train_list = []
with open('/home/xiaosucheng/Desktop/ELEC/9-3/elec-train.txt.tok','r') as f:  #注意open 和 codecs.opem
    line = f.readlines()
    for i in range(len(line)): 
        elec_train_list.append(line[i].rstrip('\n')) 
    
elec_train_label = []
with open('/home/xiaosucheng/Desktop/ELEC/9-3/elec-train.cat','r') as f:
    line = f.readlines()
    for i in range(len(line)):    
        elec_train_label.append(line[i].rstrip('\n'))
    
elec_test_list = []
with open('/home/xiaosucheng/Desktop/ELEC/9-3/elec-test.txt.tok','r') as f:  #注意open 和 codecs.opem
    line = f.readlines()
    for i in range(len(line)):    
        elec_test_list.append(line[i].rstrip('\n'))    
    
elec_test_label = []
with open('/home/xiaosucheng/Desktop/ELEC/9-3/elec-test.cat','r') as f:
    line = f.readlines() 
    for i in range(len(line)):     
        elec_test_label.append(line[i].rstrip('\n'))

# clean content
def clean_content(file_str):
    
    file_word_split = []
    count_flag = 0
    file_str = file_str.lower() # 大小写转换
    for i in range(len(file_str) - 1):

        if(file_str[i] is ' ')and(file_str[i+1] is not ' ')and(count_flag is 0):
            start_index = i+1
            count_flag = 1

        if(file_str[i] is not ' ')and(file_str[i+1] is ' ')and(count_flag is 1):
            end_index = i+1
            file_word_split.append(file_str[start_index:end_index])
            count_flag = 0
    return file_word_split

elec_train_split = []
for i in range(len(elec_train_list)):
    elec_train_split.append(clean_content(elec_train_list[i]))
    
elec_test_split  = []
for i in range(len(elec_test_list)):
    elec_test_split.append(clean_content(elec_test_list[i]))
    
# calculate vab
elec_vab_count = {}

for i in range(len(elec_train_split)):
    elec_str = elec_train_split[i]
    for word_index in range(len(elec_str)):
        if (not elec_vab_count.has_key(elec_str[word_index])):
            elec_vab_count.update({elec_str[word_index]: 1})
        else:
            elec_vab_count[elec_str[word_index]] = elec_vab_count[elec_str[word_index]] + 1
            
del_stop_word = False
if del_stop_word is True:
    # stop word list
    stop_word_list = []
    with open('/home/xiaosucheng/Desktop/ELEC/9-3/elec_stop_word.txt','r') as f:  #注意open 和 codecs.opem
        line = f.readlines()
        for i in range(len(line)):
            stop_word_list.append(line[i].rstrip('\n'))

    for i in range(len(stop_word_list)):
        if elec_vab_count.has_key(stop_word_list[i]):
            elec_vab_count.pop(stop_word_list[i])
    print 'delete stop word' 
    
    
# sort vab
elec_vab                  = {}
elec_vab_list_orgin       = elec_vab_count.keys()  # list
elec_vab_count_list_orgin = elec_vab_count.values() # list
elec_vab_list             = []
elec_vab_count_list       = []

elec_vab_index      = np.argsort(-np.array(elec_vab_count_list_orgin))  # np.argsort(-x) 降序排序
for i in range(len(elec_vab_list_orgin)):
    elec_vab.update({elec_vab_list_orgin[elec_vab_index[i]]: i + 1})    # 0 给stop_word　所以+1
    elec_vab_list.append(elec_vab_list_orgin[elec_vab_index[i]])
    elec_vab_count_list.append(elec_vab_count_list_orgin[elec_vab_index[i]])

print 'calculate vocabulary finished'

vab_size = 30000
# word2index 如果不含有该词或者index超过vab_size,则置成０
elec_train_list_index = copy.deepcopy(elec_train_split)
for i in range(len(elec_train_split)):
    elec_str = elec_train_split[i]
    for word_index in range(len(elec_str)):
        if (not elec_vab.has_key(elec_str[word_index])) or (elec_vab[elec_str[word_index]] >= vab_size):
            elec_train_list_index[i][word_index] = 0
        else:
            elec_train_list_index[i][word_index] = elec_vab[elec_str[word_index]]
print 'train word2index finished'


elec_test_list_index = copy.deepcopy(elec_test_split)
for i in range(len(elec_test_split)):
    elec_str = elec_test_split[i]
    for word_index in range(len(elec_str)):
        if (not elec_vab.has_key(elec_str[word_index])) or (elec_vab[elec_str[word_index]] >= vab_size):
            elec_test_list_index[i][word_index] = 0
        else:
            elec_test_list_index[i][word_index] = elec_vab[elec_str[word_index]]
print 'test word2index finished'

# elec_vab_list 　　　　　 词表的词
# elec_vab_count_list 　 词表词的词频
# elec_list_index 　　　 每个文档每个词的index'

import cPickle
ELEC = {}
ELEC['Vocabulary']       = elec_vab_list
ELEC['Vab_count']        = elec_vab_count_list
ELEC['Vab_Size']         = vab_size
ELEC['Label']            = ['pos','neg']
ELEC['Train_Origin']     = elec_train_list
ELEC['Train_Label']      = np.array(elec_train_label, dtype='int32') - 1
ELEC['Train_Word_Split'] = elec_train_split
ELEC['Train_Word2Index'] = elec_train_list_index
ELEC['Test_Origin']      = elec_test_list
ELEC['Test_Label']       = np.array(elec_test_label, dtype='int32') - 1
ELEC['Test_Word_Split']  = elec_test_split
ELEC['Test_Word2Index']  = elec_test_list_index
cPickle.dump(ELEC, open('./ELEC_30k.pkl','wb'))
print 'save finished'
```

```python
import numpy as np
import scipy.io as sio

ELEC['BOW'] = np.zeros([5000, 25000])
for i in range(len(ELEC['Train_Word2Index'])):
    
    elec_str = ELEC['Train_Word2Index'][i]
    for j in range(len(elec_str)):
        
        ELEC['BOW'][elec_str[j],i] = ELEC['BOW'][elec_str[j],i] + 1
        

sio.savemat('ELEC_5k.mat',{'ELEC_BOW': ELEC['BOW'], 'ELEC_Vab':ELEC['Vocabulary']})
print 'save mat finished'
```
